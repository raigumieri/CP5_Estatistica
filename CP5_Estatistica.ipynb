## Realizando as importações

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib as mpl
```

## Arquivo CSV sobre vendas de videogame



```python
game_df = pd.read_csv("vgsales.csv")

game_df
```

## **Gráfico de linha**
## Analise do valor de vendas que o XBOX 360 fez na Europa


```python
# Definindo o console
xbox = game_df[game_df["Platform"] == "X360"]

# Relacionando a venda de jogos dos EUA com os anos
Europa_sales = xbox.groupby('Year')["EU_Sales"].sum().reset_index()

Europa_sales
```

```python
# Criando o gráfico de barras
plt.figure(figsize=(8, 4))
plt.plot(Europa_sales['Year'], Europa_sales['EU_Sales'], marker='o', color='green')

# Adicionando título e rótulos
plt.title('Vendas de jogos do Xbox 360 ao longo dos anos na Europa')
plt.xlabel('Ano')
plt.ylabel('Vendas (em milhões)')
plt.grid(True)

plt.show()
```

## **Gráfico de Barras**
## Quantidade de jogos de ação feitos para o Playstation 2


```python
# Filtrando os consoles para PS2 e PS3 com o tema Ação
play = game_df[(game_df['Genre'] == 'Action') & (game_df["Platform"].isin(['PS2', 'PS3']))]

# Realizando a contagem de quantidade de jogos de ação
action_game = play['Platform'].value_counts().reset_index()
action_game.columns = ['Plataform', 'Count']

play
```

```python
# Criando o gráfico de barras
plt.figure(figsize=(8, 4))
plt.bar(action_game['Plataform'], action_game['Count'], color=['blue', 'red'])

# Adicionando título e rótulos
plt.title('Quantidade de Jogos de Ação no PS2 e PS3')
plt.xlabel('Plataforma')
plt.ylabel('Quantidade de Jogos')

plt.show()
```

## **Histograma**
## Vendas que a Nintendo fez na América do Norte

```python
# Definindo a empresa
nintendo = game_df[game_df["Publisher"] == "Nintendo"]

# Relacionando a venda de jogos da América do Norte com os anos
NA_sales = nintendo.groupby('Year')["NA_Sales"].sum().reset_index()

NA_sales
```

```python
# Criando o Histograma
plt.figure(figsize=(8, 4))
plt.hist(NA_sales['Year'], bins=10, color='purple', alpha=0.8)

# Adicionando título e rótulos
plt.title('Venda de jogos da Nintendo na América do Norte')
plt.xlabel('Ano')
plt.ylabel('Vendas (em milhões)')
plt.grid(True)

plt.show()
```

## **Gráfico de Pizza**
## Gêneros mais jogados no Japão

```python
# Filtrando dados para os jogos vendidos no Japão
JP_games = game_df[game_df["JP_Sales"] > 0]

# Agrupando por gênero e calculando as vendas totais
calc_genero = JP_games.groupby('Genre')['JP_Sales'].sum()
```

```python
# Criando o gráfico de pizza
plt.figure(figsize=(8, 8))
plt.pie(calc_genero, labels=calc_genero.index, autopct='%1.1f%%')

# Adicionando título
plt.title('Distribuição de Vendas por Gênero de Jogos no Japão')

plt.show()
```

## **Gráfico Scatter**
## Quantidade de vendas da empresa SEGA

```python
# Filtrando os dados para a empresa SEGA
sega = game_df[game_df["Publisher"] == "Sega"]

# Agrupando por ano e calculando as vendas totais
vendasT = sega.groupby('Year')['Global_Sales'].sum().reset_index()
```

```python
# Criando o gráfico Scatter
plt.figure(figsize=(8, 6))
plt.scatter(vendasT['Year'], vendasT['Global_Sales'], color='red', alpha=0.5, s=100)

# Adicionando títulos e rótulos

plt.title('Quantidade de Vendas da Sega ao Longo dos Anos')
plt.xlabel('Ano')
plt.ylabel('Vendas Globais (em milhões)')

plt.show()
```
